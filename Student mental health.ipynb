#### Exploratory Data Analysis

#### Student Conditions

```python
depressed_filter = smh['Has Depression?'] == 'Yes'
anxiety_filter = smh['Has Anxiety?'] == 'Yes'
panicking_filter = smh['Has Panic attack?'] == 'Yes'
conditions_filter = anxiety_filter | depressed_filter | panicking_filter
no_condition_filter = depressed_filter & anxiety_filter & panicking_filter
```

```python
smh['Has Condition?'] = np.where(conditions_filter, 'Yes',"No")
smh
```

```python
smh['Condition'] = ''
smh.loc[anxiety_filter, 'Condition'] += 'A'
smh.loc[depressed_filter, 'Condition'] += 'D'
smh.loc[panicking_filter, 'Condition'] += 'P'
smh.loc[no_condition_filter, 'Condition'] += 'None'
smh
```

```python
smh = smh[smh['Has Condition?'] == 'Yes']
```

```python
total_entries = len(smh)
percentage_depression = (smh['Has Depression?'].value_counts()['Yes'] / total_entries) * 100
percentage_anxiety = (smh['Has Anxiety?'].value_counts()['Yes'] / total_entries) * 100
percentage_panic_attack = (smh['Has Panic attack?'].value_counts()['Yes'] / total_entries) * 100

sizes = [percentage_depression,percentage_anxiety,percentage_panic_attack]
plt.pie(sizes, explode=(0.02,0.01,0.01), labels=['Depression','Anxiety','Panic Attack'], autopct='%1.1f%%')

plt.title('Percentage of each condition', fontsize= 12, fontweight='bold')

plt.show()
```

#### Under going treatment

```python
labels = ['Not Undergoing Treatment','Undergoing Treatment']

plt.pie(smh['Is Undergoing treatment?'].value_counts(),labels= labels, autopct='%1.1f%%', explode=(0.05,0.005),startangle=180)
plt.title('Student Undergoing Treatement', fontsize=12, fontweight='bold')
plt.show()
```

```python
sort = smh['Condition'].value_counts().index.tolist()
```

```python
sns.countplot(data=smh, x='Condition',hue='Is Undergoing treatment?', order=sort)
plt.xlabel('Condition')
plt.ylabel('Students')
plt.title('Student Undergoing Treatement', fontsize=12, fontweight='bold')
plt.show()
```

#### Student by Gender

```python
sns.countplot(data=smh, x='Condition', hue='Gender', order=sort)
plt.xlabel('Condition')
plt.ylabel('Students')
plt.title('Student by Gender', fontsize=12, fontweight='bold')
plt.show()
```
